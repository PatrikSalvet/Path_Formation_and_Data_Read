```python
def data_read(file, log_data_read = None, manual_file_format = None, data_read_option = None, setting_excel = {}, setting_table = {},
              setting_csv = {}, setting_fwf = {}):
    r"""
    Data_read function docstring, yet to be finished.
    """

#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    INITIAL FUNCTION INPUT CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°

# <-1->    <file> VARIABLE BASIC INPUT CHECK
    if type(file) not in [str, tuple, list]:
        raise Exception(f"\nThe <file> variable format type is not supported.")
    elif file in [None, [], (), {}, '']:
        raise Exception(f"\nThe <file> variable has not been defined.")

#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    MODULE IMPORT AND CONFIGURATION SETTING    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°

# <-1->    TIME MODULE IMPORT
    import time
    
# <-1->    TIMER START
    start_time = time.time()

# <-1->    OTHER MODULE IMPORT
    import numpy as np
    import pandas as pd
    import re, logging, os, copy
    import time

# <-1->    LOGGING CONFIGURATION
    logging.basicConfig(format = "%(levelname)s - %(message)s", level = logging.DEBUG)
    
# <-1->    INPUT VARIABLES FORMAT CHECK FUNCTION
    def data_read_format_check(file, log_data_read, manual_file_format, data_read_option):
        
# <-2->    ITEM UNPACKING FUNCTION
            def flatten_items(inputs):
#          inputs = True  =>  inputs = 'True'
                if isinstance(inputs, (bool, int)):
                    inputs = str(inputs)
#          inputs = 'True'  =>  inputs = ['True']
                if isinstance(inputs, str):
                    return [inputs]
#          inputs = ([[(['A']), [[('B')]]], ['C']])  =>  result = ['A', 'B', 'C']
                if isinstance(inputs, (list, tuple)):
                    result = []
                    for item in inputs:
                        result.extend(flatten_items(item))
                    return result

# <-2->    <file> VARIABLE UNPACKING
#          file = [([[['C:\\Users\\DATAXLSX.xlsx'],['C:\\Users\\data.xlsx']]])]  =>  file = ['C:\\Users\\DATAXLSX.xlsx','C:\\Users\\data.xlsx']
            file = flatten_items(file)
#          file = {'C:\\Users\\DATAXLSX.xlsx'}  =>  raise Exception 
            if not isinstance(file, (list)) or not isinstance(file[0], (str)):
                raise Exception(f"\nIncorrectly defined <file> variable.\nCheck the <file> variable format.")
#          file = ['C:\\Users\\DATAXLSX.xlsx','C:\\Users\\data.xlsx']  =>  file = [['C:\\Users\\DATAXLSX.xlsx'],['C:\\Users\\data.xlsx']]
            file = [[item] for item in file]
            
# <-2->    <log_data_read> VARIABLE UNPACKING
            if log_data_read is not None:
#          log_data_read = ([['True']])  =>  file = ['True']
                log_data_read = flatten_items(log_data_read)
#          log_data_read = {'True'}  =>  raise Exception 
                if not isinstance(log_data_read, (list)) or not isinstance(log_data_read[0], (str, bool)):
                    raise Exception(f"\nIncorrectly defined <log_data_read> variable.\nCheck the <log_data_read> variable format.")
            
# <-2->    <manual_file_format> VARIABLE UNPACKING
            if manual_file_format is not None:          
#          manual_file_format = ([['.xlsx']])  =>  manual_file_format = ['.xlsx']           
                manual_file_format = flatten_items(manual_file_format)          
#          manual_file_format = {'.xlsx'}  =>  raise Exception 
                if not isinstance(manual_file_format, (list)) or not isinstance(log_data_read[0], (str)):
                    raise Exception(f"\nIncorrectly defined <manual_file_format> variable.\nCheck the <manual_file_format> variable format.")
                          
# <-2->    <data_read_option> VARIABLE UNPACKING
            if data_read_option is not None:                
#          data_read_option = ([['pd.read_excel']])  =>  data_read_option = ['pd.read_excel']
                data_read_option = flatten_items(data_read_option)        
#          data_read_option = {'pd.read_excel'}  =>  raise Exception          
                if not isinstance(data_read_option, (list)) or not isinstance(data_read_option[0], (str)):
                    raise Exception(f"\nIncorrectly defined <data_read_option> variable.\nCheck the <data_read_option> variable format.")            
                          
# <-2->    <data_read_format_check> FUNCTION OUTPUT VARIABLE DEFINITION
            return file, log_data_read, manual_file_format, data_read_option

# <-1->    <data_read_format_check> FUNCTION CALL
    format_check = True
    if format_check:
        file, log_data_read, manual_file_format, data_read_option = data_read_format_check(file, log_data_read, manual_file_format, data_read_option)

# <-1->    <log_data_read> VARIABLE LIST UNPACKING
#          log_data_read = ['True']  =>  log_data_read = 'True'      log_data_read = 'True'  <=>  log_data_read = 'True'
    if type(log_data_read) is not None and (type(log_data_read) == list or type(log_data_read) == tuple) and len(log_data_read) > 0:
        if len(log_data_read) > 1:
            logging.warning(f"Multiple items have been found in the <log_data_read> variable, only the first item will be used.")
        log_data_read = log_data_read[0]
        
# <-1->    <log_data_read> VARIABLE INCORRECT FORMAT TYPE NONE ASSIGNMENT
#          log_data_read = ('True','True')  =>  log_data_read = None      log_data_read = 'True','True'  =>  log_data_read = None
    elif type(log_data_read) == type([]) or type(log_data_read) == type(()):
        log_data_read = None

# <-1->    <log_data_read> VARIABLE INPUT RECOGNITION AND BOOLEAN ASSIGNMENT
    if locals().get('log_data_read') is not None and type(log_data_read) is not bool:
        try:
#          log_data_read = 'abcTrue23False'  =>  len(log_data_read_yes) != 0        log_data_read = 'abc023'  =>  len(log_data_read_yes) == 0
            log_data_read_yes = re.compile(r'(1|one|true|yes|on|enabled|enable)').findall(log_data_read.lower())
#          log_data_read = 'abcTrue23False'  =>  len(log_data_read_no) != 0        log_data_read = 'abc123'  =>  len(log_data_read_no) == 0
            log_data_read_no = re.compile(r'(0|two|false|no|off|none|disabled|disable)').findall(log_data_read.lower())
            if (len(log_data_read_yes) != 0 ) and (len(log_data_read_no) == 0):
                log_data_read = True      
            elif (len(log_data_read_yes) == 0 ) and (len(log_data_read_no) != 0):
                log_data_read = False
            elif (len(log_data_read_yes) != 0 ) and (len(log_data_read_no) != 0):
                logging.warning(f"Incorrectly defined <log_data_read> variable. The logging has been enabled by default.")
                log_data_read = True
            elif (len(log_data_read_yes) == 0 ) and (len(log_data_read_no) == 0):
                if log_data_read in ['', ' ']:
                    logging.warning(f"The <log_data_read> variable has not been defined. The logging has been enabled by default.")
                    log_data_read = True
                else:
                    logging.warning(f"Incorrectly defined <log_data_read> variable. The logging has been enabled by default.")
                    log_data_read = True 
        except:
            logging.warning(f"Incorrectly defined <log_data_read> variable. Logging has been enabled by default.")
            log_data_read = True
        
# <-1->    LOGGING ACTIVATION
    if log_data_read is not None and log_data_read == True:
        logging.disable(logging.NOTSET)
        logging.info(f"Logging has been enabled for the <data_read> function.")
    
# <-1->    <log_data_read> VARIABLE BOOLEAN ASSIGNMENT BASED ON GLOBAL LOG VARIABLE
    elif (locals().get('log_data_read') is None) and (globals().get('log_global') is not None):
        log_data_read = log_global
        if log_data_read == True:
            logging.info(f"Logging has been enabled based on the previously called function.")
        else:
            logging.info(f"Incorrectly defined <log_data_read> variable. Logging has been enabled by default.")

# <-1->      UNKNOWN <log_data_read> VARIABLE DEFAULT VALUE ASSIGNMENT
    elif (locals().get('log_data_read') is None) and (globals().get('log_global') is None):
        log_data_read = True
        logging.info(f"The <log_data_read> variable is not defined. Logging has been enabled by default.")
    
# <-1->    LOGGING ACTIVATION
    if log_data_read == True:
        logging.disable(logging.NOTSET)
    
# <-1->    LOGGING DEACTIVATION
    elif log_data_read == False:
        logging.disable(logging.CRITICAL)

# <-1->    FUNCTION INPUT DISPLAY
    logging.info(f"The <data_read> function has been called.")
    
# <-1->    <warnings> VARIABLE CREATION AND VALUE RESET
    warnings = False
    
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    MANUAL FILE FORMAT SETTING    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°

# <-0->    MANUAL FILE FORMAT RECOGNITION SECTION
    if (locals().get('manual_file_format') not in [None, [], (), {}, '', False]):

# <-1->    <supported_files> and <file_format> VARIABLE CREATION
        supported_files = ['xlsx', 'txt', 'csv']
        logging.info(f"The supported file formats are: {supported_files}") 
        file_format = copy.deepcopy(manual_file_format)

# <-1->    <file_format> VARIABLE LIST CONVERSION
#          file_format = ('xlsx')  =>  file_format = ['xlsx']        file_format = 'xlsx'  =>  file_format = ['xlsx']
        if type(file_format) == str:
            file_format = [file_format]
#          file_format = ('xlsx', 'txt')  OR  file_format = 'xlsx', 'txt'  =>  file_format = ['xlsx', 'txt']
        elif type(file_format) == tuple:
            file_format = [*file_format]
        logging.info(f"The entered file formats are: {file_format}")

# <-1->    <file_format> VARIABLE EMPTY SPACE AND SYMBOL FILTRATION
#          file_format = ['x*l?!s x+']  =>  file_format = ['xlsx']
        for index, file_format_i in enumerate(file_format):
            filtered_symbols = re.compile(r"[^\s\^\?\!\#\*\%\/\\\<\>\{\}\&\$\'\"\:\;\|\@\+\-\`\[\]\(\)\.\_]*").findall(file_format_i)
            file_format[index] = (''.join([i for i in filtered_symbols])).lower()

# <-1->    <file_format> VARIABLE DUPLICATE FILTRATION
#          file_format = ['xlsx', 'xlsx', 'txt']  =>  file_format = ['xlsx', 'txt']
        unique_file_format = []
        [unique_file_format.append(i) for i in file_format if i not in unique_file_format]

# <-1->    <file_format> VARIABLE UNSUPPORTED FILE FORMAT FILTRATION
#          supported_files = ['xlsx', 'txt', 'csv']   and   file_format = ['xlsx', 'txt', 'ABC123']  =>  file_format = ['xlsx', 'txt']
        file_format = []
        [file_format.append(file_format_i) for file_format_i in unique_file_format if file_format_i in supported_files]
        if file_format == []:
            logging.warning(f"File format filtering has been performed, none of the file formats have been recognized.")
            warnings = True
        else:
            logging.info(f"File format filtering has been performed, the following file formats have been recognized: {file_format}")

# <-1->    <file_format> VARIABLE UNSUPPORTED FILE FORMATS DETECTED, AUTOMATIC FILE FORMAT FILTRATION ENABLED
        if len(unique_file_format) > len(file_format):
#          file_format = ['xlsx', 'ABC123']  AND  supported_files = ['xlsx']  =>  file_format_difference = ['ABC123'] AND manual_file_format = None
            file_format_difference = list(set(unique_file_format).symmetric_difference(set(file_format)))
            manual_file_format = None
            logging.warning(f"The following file formats are not supported: {file_format_difference}")
            logging.warning(f"The file format will be determined using automatic file format recognition function.")
            warnings = True

# <-1->    <file_format> VARIABLE NO VALID FILE FORMAT RECOGNIZED, AUTOMATIC FILE FORMAT FILTRATION ENABLED
        if len(file_format) == 0 and manual_file_format != None:
#          file_format = ['']  =>  manual_file_format = None
            manual_file_format = None
            logging.warning(f"None of the file formats specified in the <manual_file_format> variable were valid.")
            logging.warning(f"The file format will be determined using automatic file format recognition function.")
            warnings = True

# <-1->    <file_format> VARIABLE MULTIPLE VALID FILE FORMATS RECOGNIZED, AUTOMATIC FILE FORMAT FILTRATION ENABLED
#          file_format = ['txt','xlsx']  =>  manual_file_format = None
        elif len(file_format) > 1 and manual_file_format != None:  
            manual_file_format = None
            logging.warning(f"Multiple file formats have been recognized; however, only one can be used at a time.")
            logging.warning(f"The file format will be determined using automatic file format recognition function.")
            warnings = True

# <-1->    <file_format> CORRECT INPUT DISPLAY
        if manual_file_format not in [None, [], (), {}, '', False]:
            manual_file_format = 'Final file format stored in the file_format variable'
#          file_format = ['xlsx']  =>  file_format = ['.xlsx']
            file_format[0] = '.' + file_format[0]
            logging.info(f"File format operations finished. The file format has been set to: {file_format}")

#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    AUTOMATIC FILE FORMAT RECOGNITION    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°

# <-0->    AUTOMATIC FILE FORMAT RECOGNITION SECTION
    if (locals().get('manual_file_format') in [None, [], (), {}, '', False]):
        logging.info(f'The automatic file format recognition function will be executed.')
    
# <-1->    NEW VARIABLE CREATION
        supported_file_formats = ['.xlsx', '.txt', '.csv']
        default_file_format = ['.xlsx']
        file_format, unique_file_format, unique_formats = [], [], []

# <-1->    <file> VARIABLE POSTFIX IDENTIFICATION
#          file = [['C:\\Users\\Folder\\DATAXLSX.xlsx'],['C:\\Users\\Folder\\DATAXLSX.txt']]   =>   file_format = ['.xlsx', '.txt']
        [file_format.append(os.path.splitext(file_item_list[0])[-1].lower()) for file_item_list in file]

# <-1->    <file_format> VARIABLE EMPTY SPACE AND SYMBOL FILTRATION
#          file_format = ['x*l?!s x+']  =>  file_format = ['xlsx']
        for index, file_format_i in enumerate(file_format):
            filtered_symbols = re.compile(r"[^\s\^\?\!\#\*\%\/\\\<\>\{\}\&\$\'\"\:\;\|\@\+\-\`\[\]\(\)\_]*").findall(file_format_i)
            file_format[index] = (''.join([i for i in filtered_symbols])).lower()

# <-1->    <file_format> VARIABLE DUPLICATE POSTFIX FILTRATION
#          file_format = ['.xlsx', '.txt', '', '.xlsx']   =>   unique_file_format = ['.xlsx', '.txt']
        [unique_file_format.append(item) for item in file_format if item not in unique_file_format and item != '']
        logging.info(f'The unique file formats found in the <file> variable are: {unique_file_format}')

# <-1->    <unique_file_format> VARIABLE UNSUPPORTED FILE FORMAT FILTRATION
#          unique_file_format = ['.xlsx', '.csv']  AND  supported_file_formats = ['.xlsx', '.txt']   =>   unique_formats = ['.xlsx']
        [unique_formats.append(file_item) for file_item in unique_file_format if file_item in supported_file_formats]

# <-1->    ALL IDENTIFIED FILE FORMATS SUPPORTED MESSAGE
        if len(unique_file_format) == len(unique_formats) != 0:
            logging.info(f'All files found in the <file> variable are in the supported file formats.')
    
# <-1->    SOME IDENTIFIED FILE FORMATS NOT SUPPORTED MESSAGE
        elif len(unique_file_format) != len(unique_formats) != 0:
#          unique_file_format = ['.xlsx', '.csv']  AND  unique_formats = ['.xlsx']   =>   format_difference = ['.csv']
            format_difference = list(set(unique_file_format).symmetric_difference(set(unique_formats)))
            logging.warning(f'The following file formats found in the <file> variable are not supported: {format_difference}')
            warnings = True
            
# <-1->    NO FILE FORMAT FOUND MESSAGE
        elif unique_formats == [] and unique_file_format == ['']:
            logging.warning(f'No file formats in the <file> variable have been recognized.')
            logging.warning(f'The file format has been changed to the following default value: {default_file_format}')
            warnings = True
            file_format = default_file_format

# <-1->    NO SUPPORTED FILE FORMAT FOUND MESSAGE
        elif len(unique_formats) == 0:
            logging.warning(f'None of the file formats found in the <file> variable are supported.')
            logging.warning(f'The file format has been changed to the following default value: {default_file_format}')
            warnings = True
            file_format = default_file_format
    
# <-1->     SUCCESSFUL FILE FORMAT RECOGNITION MESSAGE
        if len(unique_formats) == 1 == len(unique_file_format):
            file_format = unique_formats
            logging.info(f'The file format has been successfully recognized as: {file_format}')
    
# <-1->     MULTIPLE FILE FORMATS RECOGNIZED MESSAGE
        elif len(unique_formats) == 1 and len(unique_file_format) > 1:
            file_format = unique_formats
            logging.warning(f"Of the recognized file formats, only the '{file_format[0]}' file format is supported.")
            logging.warning(f'The file format has been set to: {file_format}')
            warnings = True

# <-1->     MULTIPLE FILE FORMATS RECOGNIZED MESSAGE
        elif len(unique_formats) > 1:
            file_format = [unique_formats[0]]
            logging.warning(f'The following recognized file formats are supported, but only one can be used at a time: {unique_formats}')
            logging.warning(f'The file format has been set to: {file_format}')
            warnings = True

#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    FILE PATH CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°  

# <-1->    <file> VARIABLE LIST FORMAT CONSTRUCTION
#          file = 'C:\\Users\\Folder\\DATA.txt'  =>  file = ['C:\\Users\\Folder\\DATA.txt']
    if type(file) == str:
        file = [file]
#          file = ['C:\\Users\\DATA.txt', 'C:\\Users\\N03.xlsx']  =>  file = [['C:\\Users\\DATA.txt'], ['C:\\Users\\N03.xlsx']]
    if type(file[0]) != list:
        file = [[item] for item in file]

# <-1->    <file> VARIABLE UNSUPPORTED SYMBOL FILTRATION
#          file = [[' C:\\Use*?rs\\DAT?A.txt\/\ '], [' \C:\\Us?*<ers\\N0<3.xls<x ']]  =>  file = [['C:\\Users\\DATA.txt'], ['C:\\Users\\N03.xlsx']]
    for i, item in enumerate(file):
        item[0] = item[0].strip(' /\\')
        item[0] = re.compile(r"[^\*\?\"\|\<\>]*").findall(item[0])
        item[0] = (''.join(item[0]))
        file[i] = [item[0]]
    logging.info(f'File path punctuation symbol filtration has been performed.')
    logging.info(f'The searched file paths are: {file}')

# <-1->    NEW VARIABLE CREATION
    file_found, file_notfound = [], []

# <-1->    <file> VARIABLE ITEM SEARCH
    for item in file:
#          file = [['C:\\Users\\DATA.txt']]  AND  os.listdir() = ['N02.xlsx', DATA.txt]  =>  file_found = [['C:\\Users\\DATA.txt']]
        if os.path.exists(item[0]):
#          item = [['C:\\Users\\DATA.txt']]   =>  os.getcwd() = 'C:\\Users'
            current_path = os.path.dirname(os.path.abspath(item[0]))
            os.chdir(current_path)
#          file = [['C:\\Users\\DATA.txt'], [N02.txt]] AND os.listdir() = ['DATA.txt', 'N02.txt']  =>  file_found = [['C:\\Users\\DATA.txt'], ['C:\\Users\\N02.txt]]
            if not os.path.isabs(item[0]):
                file_found.append([os.getcwd() + os.sep + item[0]])
            else:
                file_found.append(item)
#          file = [['C:\\Users\\DATA.txt']]  AND  os.listdir() = ['N02.xlsx', N03.txt]  =>  file_notfound = [['C:\\Users\\DATA.txt']]
        else:
            file_notfound.append(item)

# <-1->    <file> VARIABLE ALL FILES FOUND MESSAGE
    if len(file_notfound) == 0 and len(file_found) != 0:
        logging.info(f'All files specified in the <file> variable have been found.')
    
# <-1->    <file> VARIABLE SOME FILES NOT FOUND MESSAGE
    elif len(file_notfound) != 0 and len(file_found) != 0:
#          file_notfound = [['C:\\Users\\DATA1.txt'],['C:\\Users\\DATA2.txt']]  =>  item_notfound = [['DATA1.txt'], ['DATA2.txt']]
        item_notfound = [[os.path.basename(item[0])] for item in file_notfound]
        logging.warning(f'Some of the files specified in the <file> variable could not be found.')
        logging.warning(f'The following files could not be found: {item_notfound}.')
        warnings = True

# <-1->    <file> VARIABLE NO FILES FOUND ERROR
    elif len(file_found) == 0:
        raise Exception(f'\nNo files specified in the the <file> variable have been found.')

# <-1->   <file> VARIABLE FOUND FILES MESSAGE
#          file_found = [['C:\\Users\\DATA1.txt'],['C:\\Users\\DATA2.txt']]  =>  item_found = [['DATA.txt'], ['N03.xlsx']]
    item_found = [[os.path.basename(item[0])] for item in file_found]
    logging.info(f'The following files have been found: {item_found}')

# <-1->   <file> VARIABLE UNSUPPORTED FILE FILTRATION
#          file_format = ['.xlsx']  AND  item_found = [['DATA.txt'], ['N03.xlsx']]  =>  item_found_currentformat = [['N03.xlsx']]
    item_found_currentformat = [item for item in item_found if file_format[0].lower() in item[0].lower()]

# <-1->   <file> VARIABLE FILE FORMAT ERROR
    if len(item_found_currentformat) == 0:
        raise Exception(f'\nNone of the detected files are in the currently used file format.')
    
# <-1->   <file> VARIABLE CORRECT FILE FORMAT MESSAGE
    elif len(item_found_currentformat) > 0 and len(item_found_currentformat) == len(item_found):
        logging.info(f'All found files are in the currently used file format.')
    
# <-1->   <file> VARIABLE WRONG FILE FORMAT WARNING MESSAGE
    if len(item_found_currentformat) > 0 and len(item_found_currentformat) != len(item_found):
#          item_found = [['DATA.txt'], ['N03.xlsx']]  =>  flat_item_found = ['DATA.txt', 'N03.xlsx']
        flat_item_found = [item[0] for item in item_found]
#          item_found_currentformat = [['N03.xlsx'], ['N02.xlsx']]  =>  flat_item_currentformat = ['N03.xlsx', 'N02.xlsx']
        flat_item_currentformat = [item[0] for item in item_found_currentformat]
#          flat_item_found = ['DATA.txt', 'N03.xlsx']  AND  flat_item_currentformat = ['N03.xlsx']  =>  file_item_difference = ['DATA.txt']
        file_item_difference = list(set(flat_item_currentformat).symmetric_difference(set(flat_item_found)))
        logging.warning(f'Some of the found files are not in the current file format.')
        logging.warning(f'The following files could not be loaded: {file_item_difference}')
        warnings = True
        
# <-1->   FINAL FILE PATH VARIABLE TO READ CONSTRUCTION
#          file_found = [['C:\\DATA1.XlsX'],['C:\\DATA2.txt']]  AND  file_format = ['.xlsx']  =>  file_to_read = [['C:\\DATA1.xlsx']]
    file_to_read = [[os.path.splitext(item[0])[0] + os.path.splitext(item[0])[1].lower()] for item in file_found if item[0].lower().endswith(file_format[0].lower())]
    logging.info(f'The following files will be loaded: {file_to_read}')
        
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    DATA READ OPTION INPUT CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°  

# <-1->    <default_data_read_option> VARIABLE RESET
    default_data_read_option = None

# <-1->    <supported_data_read_options> VARIABLE CREATION
    supported_data_read_options = ['pd.read_excel', 'pd.read_table', 'pd.read_csv', 'pd.read_fwf']
    logging.info(f"The supported data read options are: {supported_data_read_options}")

# <-1->    <data_read_option> VARIABLE INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          data_read_option = False  =>  data_read_option = []
    if (locals().get('data_read_option') in [False, True, None]):
        data_read_option = []

# <-1->    <data_read_option> VARIABLE LIST CONVERSION
#          data_read_option = ('pd.read_excel')  OR  data_read_option = 'pd.read_excel'  =>  data_read_option = ['pd.read_excel']
    if type(data_read_option) == str:
        data_read_option = [data_read_option]
#          data_read_option = ('pd.read_excel', 'pd.read_table')  =>  data_read_option = ['pd.read_excel', 'pd.read_table']
    elif type(data_read_option) == tuple:
        data_read_option = [*data_read_option]
    logging.info(f"The entered data read options are: {data_read_option}")

# <-1->    <data_read_option> VARIABLE EMPTY SPACE AND SYMBOL FILTRATION
#          data_read_option = ['pd.rea?!*d_exc+el']  =>  data_read_option = ['pd.read_excel']
    for index, data_read_option_i in enumerate(data_read_option):
        filtered_symbols = re.compile(r"[^\s\^\?\!\#\*\%\/\\\<\>\{\}\&\$\'\"\:\;\|\@\+\-\`\[\]\(\)]*").findall(data_read_option_i)
        data_read_option[index] = (''.join([i for i in filtered_symbols])).lower()

# <-1->    <data_read_option> VARIABLE DUPLICATE FILTRATION
#          data_read_option = ['pd.read_excel', 'pd.read_excel']  =>  data_read_option = ['pd.read_excel']
    unique_data_read_option = []
    [unique_data_read_option.append(i) for i in data_read_option if i not in unique_data_read_option]

# <-1->    <data_read_option> VARIABLE UNSUPPORTED DATA READ OPTION FILTRATION
#          supported_data_read_options = ['pd.read_excel'] AND data_read_option = ['xlsx', 'pd.read_excel']  =>  data_read_option = ['pd.read_excel']
    data_read_option = []
    [data_read_option.append(data_read_option_i) for data_read_option_i in unique_data_read_option if data_read_option_i in supported_data_read_options]
    if len(data_read_option) > 0:
        logging.info(f"The following data read options have been recognized: {data_read_option}")
    elif len(data_read_option) == 0:
        logging.warning(f"None of the entered data read options have been recognized.") 
        warnings = True
        
# <-1->    <data_read_option> VARIABLE UNSUPPORTED OPTIONS DETECTED MESSAGE PRINT
    if len(unique_data_read_option) > len(data_read_option):
#          data_read_option = ['pd.read_excel', 'ABC123'] AND supported_data_read_options = ['pd.read_excel'] => data_read_option_difference = ['ABC123']
        data_read_option_difference = list(set(unique_data_read_option).symmetric_difference(set(data_read_option)))
        logging.warning(f"The following data read options are not supported: {data_read_option_difference}")
        logging.warning(f"The unsupported data read options will be ignored.")
        warnings = True

# <-1->    <data_read_option> VARIABLE NO VALID DATA READ OPTION RECOGNIZED MESSAGE PRINT
    if len(data_read_option) == 0:
#          data_read_option = ['']  =>  manual_file_format = None
        logging.warning(f"None of the data read options specified in the <data_read_option> variable were valid.")
        logging.warning(f"The data read option will be set based on the <file_format> variable default value.")
        default_data_read_option = True
        warnings = True

# <-1->    <data_read_option> VARIABLE MULTIPLE VALID DATA READ OPTIONS RECOGNIZED MESSAGE PRINT
    elif len(data_read_option) > 1:  
        logging.warning(f"Multiple valid data read options have been recognized; however, only one can be used at a time.")
        logging.warning(f"The data read option will be set based on the <file_format> variable default value.")
        default_data_read_option = True
        warnings = True

# <-1->    <data_read_option> INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
    if locals().get('default_data_read_option') == True:
        if file_format == ['.xlsx']:
            data_read_option = ['pd.read_excel']
        elif file_format == ['.txt']:
            data_read_option = ['pd.read_fwf']
        elif file_format == ['.csv']:
            data_read_option = ['pd.read_csv']
        else:
            data_read_option = ['pd.read_table']
        
# <-1->    FINAL <data_read_option> VARIABLE DISPLAY
    if default_data_read_option == True:
        logging.warning(f"The data read option has been set to: {data_read_option}")
        warnings = True
    else:
        logging.info(f"The data read option has been set to: {data_read_option}")
    
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    EXCEL DATA READ FUNCTION INPUT CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°  

# <-0->    <setting_excel> VARIABLE SETTING SECTION
    if data_read_option == ['pd.read_excel']:

# <-1->    DICTIONARY UNPACKING FUNCTION DEFINITION
#          item = [[(([{'sheet_name': 0}]))]]   =>   dict_unpack(item)   =>   item = {'sheet_name': 0}
        def dict_unpack(item):
            if isinstance(item, list):
                return dict_unpack(item[0])
            elif isinstance(item, dict):
                return item

# <-1->    <setting_excel> VARIABLE UNPACKING
#          setting_excel = [[(([{'sheet_name': 0}]))]]   =>   setting_excel = {'sheet_name': 0}
        setting_excel = dict_unpack(setting_excel)

# <-1->    INCORRECT <setting_excel> VARIABLE TYPE
#          setting_excel = [{'sheet_name': 0, 'displayed_decimals': 5, 'dtype': 'np.float64'}]
        if type(setting_excel) != dict:
            raise Exception(f"\nIncorrectly defined <setting_excel> variable.\nThe <setting_excel> variable must be a dictionary.")

# <-1->    DICTIONARY ITEM UNPACKING FUNCTION DEFINITION
#          item = {'dict_key': ([(['value'])])}   =>   unpack_list(item)   =>   item = {'dict_key': 'value'}
        def unpack_list(item):
            if isinstance(item, list):
                return unpack_list(item[-1])
            else:
                return item

# <-1->    <setting_excel> VARIABLE DICTIONARY VALUE LIST UNPACKING
#          setting_excel = {'displayed_decimals': ([([5,3,2])])}   =>   setting_excel = {'displayed_decimals': 2}
        items_to_unpack = ['displayed_decimals', 'dtype', 'engine', 'skiprows', 'skipfooter']
        for key, value in setting_excel.items():
            if key.lower() in items_to_unpack:
                setting_excel[key] = unpack_list(value)

# <-1->    DUMMY DICTIONARY AND ACCEPTED DICTIONARY KEY ITEMS VARIABLE CREATION
        setting_excel_accepted_keys = ['sheet_name', 'displayed_decimals', 'dtype', 'skiprows', 'skipfooter', 'usecols', 'engine']
        unknown_keys = {}

# <-1->    <setting_excel> VARIABLE LOWER CASE CONVERTION
#          setting_excel = {'shEEt_naMe': 'A0', 'dispLayed_dEciMals': 5}   =>   setting_excel = {'sheet_name': 'A0', 'displayed_decimals': 5}
        setting_excel = {key.lower(): value.lower() if isinstance(value, str) and key.lower() not in ['sheet_name', 'usecols'] else value for key, value in setting_excel.items()}

# <-1->    DICTIONARY DUPLICATE FILTRATION FUNCTION DEFINITION
#          input_dict = {'skiprows': 0, 'skiprows': 0}   =>   input_dict = {'skiprows': 0}
        def remove_duplicate_dict(input_dict):
            dict_result = {}
            for key, value in input_dict.items():
                dict_result.setdefault(key, value)
            return dict_result

# <-1->    <setting_excel> VARIABLE DUPLICATE FILTRATION
#          setting_excel = {'sheet_name': '0', 'sheet_name': '0'}   =>   setting_excel = {'sheet_name': '0'}
        setting_excel = remove_duplicate_dict(setting_excel)    

# <-1->    <setting_excel> VARIABLE UNKNOWN KEY SEARCH
#          setting_excel = {'shieeet_name': '0', 'sheet_name': '0'} AND setting_excel_accepted_keys = ['sheet_name']   =>   unknown_keys = {'shieeet_name': 0'}
        for key, value in setting_excel.items():
            if key.lower() not in setting_excel_accepted_keys:
                unknown_keys.setdefault(key, value)

# <-1->    <setting_excel> VARIABLE UNKNOWN ITEM FILTRATION
        if unknown_keys != {}:
            logging.warning(f"The following <setting_excel> inputs have not been recognized: {unknown_keys}")
#          setting_excel = {'sheet_name': 0, 'astype': 'np.float64'} AND unknown_keys = {'astype': 'np.float64'}   =>   setting_excel = {'sheet_name': 0}
            setting_excel = {key: value for key, value in setting_excel.items() if key not in unknown_keys.keys()}
            warnings = True
        elif unknown_keys == {} and len(setting_excel) > 0:
            logging.info(f"All <setting_excel> inputs have been recognized.")
    
# <-1->    <setting_excel> VARIABLE UNDEFINED ITEMS NONE VALUE ASSIGNMENT
#          setting_excel = {'dtype': 'float64'}  AND  setting_excel_accepted_keys = ['engine', 'dtype']   =>   setting_excel = {'dtype': 'float64', 'engine': None}
        for item in setting_excel_accepted_keys:
            if item not in setting_excel.keys():
                setting_excel.setdefault(item, None)

# <-1->    <setting_excel> 'dtype' DEFAULT VALUE CREATION
        default_dtype = np.float64

# <-1->    <setting_excel> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'dtype': None}   =>   setting_excel = {'dtype': 'float64'}
        if setting_excel.get('dtype') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_excel['dtype'] = default_dtype
            logging.warning(f"The <setting_excel> dtype has not been defined.")
            logging.warning(f"The <setting_excel> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_excel> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_excel={}  =>  setting_excel = {'dtype':'np.float64'}   |  setting_excel = {'dtype':'np.float16'}  <=>  setting_excel = {'dtype':'np.float16'}
        add_setting_excel = {'dtype': default_dtype}
        [setting_excel.setdefault(key, value) for key, value in add_setting_excel.items()]

# <-1->    <setting_excel> 'dtype' KEY INPUT STRING MODIFICATION
#          setting_excel = {'dtype':'np.float64'} => setting_excel = {'dtype':'float64'}
        if type(setting_excel['dtype']) == str:
            setting_excel['dtype'] = setting_excel['dtype'].replace('np.','')

# <-1->    <setting_excel> 'dtype' KEY NUMPY VALUE ASSIGNMENT
#          setting_excel = {'dtype': 'float64'}   =>   setting_excel = {'dtype': np.float64}
            try:
                setting_excel['dtype'] = getattr(np, setting_excel['dtype'])
#          setting_excel = {'dtype': 'float1000'}   =>   setting_excel = {'dtype': None}
            except:
                setting_excel['dtype'] = None

# <-1->    <setting_excel> 'dtype' KEY INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'dtype': None}   =>   setting_excel = {'dtype': 'float64'}
        if setting_excel['dtype'] not in [np.float64, np.float32, np.float16, np.int64, np.int32, np.int16,\
                                    np.uint64, np.uint32, np.uint16, int, float]:
            setting_excel['dtype'] = default_dtype
            logging.warning(f"Incorrectly defined <setting_excel> dtype.")
            logging.warning(f"The <setting_excel> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_excel> 'dtype' KEY ASSIGNED VALUE DISPLAY
        logging.info(f"The <setting_excel> dtype is set to: '{setting_excel['dtype']}'")
    
# <-1->    <accepted_engine> VARIABLE CREATION
        accepted_engine = ['xlrd', 'openpyxl', 'odf', 'pyxlsb']

# <-1->    <setting_excel> 'engine' DEFAULT VALUE ASSIGNMENT
#          setting_excel = {}   =>   setting_excel = {'engine': 'openpyxl'}
        if setting_excel.get('engine') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_excel['engine'] = 'openpyxl'
            logging.info(f"The <setting_excel> engine has been set to the following default value: '{setting_excel['engine']}'")

# <-1->    <setting_excel> 'engine' CORRECT INPUT MESSAGE
        elif setting_excel.get('engine') in accepted_engine:
            logging.info(f"The <setting_excel> engine is set to: '{setting_excel['engine']}'")

# <-1->    <setting_excel> 'engine' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'engine': ''}   =>   setting_excel = {'engine': 'openpyxl'}
        elif setting_excel.get('engine') not in accepted_engine:
            setting_excel['engine'] = 'openpyxl'
            logging.info(f"The accepted <setting_excel> engine values are: {accepted_engine}")
            logging.warning(f"Incorrectly defined <setting_excel> engine.")
            logging.warning(f"The <setting_excel> engine has been set to the following default value: '{setting_excel['engine']}'")
            warnings = True

# <-1->    <setting_excel> 'skiprows' STRING TO INTEGER CONVERSION
#          setting_excel = {'skiprows': '0'}   =>   setting_excel = {'skiprows': 0}
        try:
            if type(setting_excel['skiprows']) == str:
                setting_excel['skiprows'] = int(setting_excel['skiprows'])
#          setting_excel = {'skiprows': 'one'}   <=>   setting_excel = {'skiprows': 'one'}
        except:
            pass

# <-1->    <setting_excel> 'skiprows' DEFAULT VALUE ASSIGNMENT
#          setting_excel = {}   =>   setting_excel = {'skiprows': 0}
        if setting_excel.get('skiprows') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_excel['skiprows'] = 0
            logging.info(f"The <setting_excel> skiprows has been set to the following default value: {setting_excel['skiprows']}")
            logging.info(f"No initial rows have been skipped in the data loading process.")

# <-1->    <setting_excel> 'skiprows' CORRECT INPUT MESSAGE
        elif type(setting_excel.get('skiprows')) == int:
            logging.info(f"The <setting_excel> skiprows is set to: {setting_excel['skiprows']}")
            logging.info(f"The number of skipped initial rows in the data loading process was: {setting_excel['skiprows']}")

# <-1->    <setting_excel> 'skiprows' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'skiprows': 'one'}   =>   setting_excel = {'skiprows': 0}
        elif type(setting_excel.get('skiprows')) != int:
            setting_excel['skiprows'] = 0
            logging.warning(f"The <setting_excel> skiprows value must be an integer.")
            logging.warning(f"The <setting_excel> skiprows has been set to the following default value: {setting_excel['skiprows']}")
            logging.warning(f"No initial rows have been skipped in the data loading process.")
            warnings = True

# <-1->    <setting_excel> 'skipfooter' STRING TO INTEGER CONVERSION
#          setting_excel = {'skipfooter': '0'}   =>   setting_excel = {'skipfooter': 0}
        try:
            if type(setting_excel['skipfooter']) == str:
                setting_excel['skipfooter'] = int(setting_excel['skipfooter'])
#          setting_excel = {'skipfooter': 'one'}   <=>   setting_excel = {'skipfooter': 'one'}
        except:
            pass

# <-1->    <setting_excel> 'skipfooter' DEFAULT VALUE ASSIGNMENT
#          setting_excel = {}   =>   setting_excel = {'skipfooter': 0}
        if setting_excel.get('skipfooter') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_excel['skipfooter'] = 0
            logging.info(f"The <setting_excel> skipfooter has been set to the following default value: {setting_excel['skipfooter']}")
            logging.info(f"No end rows have been skipped in the data loading process.")

# <-1->    <setting_excel> 'skipfooter' CORRECT INPUT MESSAGE
        elif type(setting_excel.get('skipfooter')) == int:
            logging.info(f"The <setting_excel> skipfooter is set to: {setting_excel['skipfooter']}")
            logging.info(f"The number of skipped end rows in the data loading process was: {setting_excel['skipfooter']}")

# <-1->    <setting_excel> 'skipfooter' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'skipfooter': 'one'}   =>   setting_excel = {'skipfooter': 0}
        elif type(setting_excel.get('skipfooter')) != int:
            setting_excel['skipfooter'] = 0
            logging.warning(f"The <setting_excel> skipfooter value must be an integer.")
            logging.warning(f"The <setting_excel> skipfooter has been set to the following default value: {setting_excel['skipfooter']}")
            logging.warning(f"No end rows have been skipped in the data loading process.")
            warnings = True

# <-1->    <setting_excel> 'usecols' TUPLE TO LIST CONVERSION
#          setting_excel = {'usecols': (1,2,3)}   =>   setting_excel = {'usecols': [1, 2, 3]}
        try:
            if type(setting_excel.get('usecols')) in [tuple]:
                setting_excel['usecols'] = list(setting_excel['usecols'])

# <-1->    <setting_excel> 'usecols' STRING OR INTEGER TO LIST CONVERSION
#          setting_excel = {'usecols': 1}   =>   setting_excel = {'usecols': [1]}
            elif type(setting_excel.get('usecols')) in [str, int]:
                setting_excel['usecols'] = [setting_excel['usecols']]

# <-1->    <setting_excel> 'usecols' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_excel = {}   =>   setting_excel = {'usecols': None}
            if setting_excel.get('usecols') in [[''], [' '], [], (), {}, None, '', ' ']:
                if setting_excel.get('usecols') in [None]:
                    logging.info(f"The columns of each excel file to load are specified as: {setting_excel['usecols']}")
                else:
                    setting_excel['usecols'] = None
                    logging.info(f"The <setting_excel> usecols has been set to the following default value: '{setting_excel['usecols']}'")
                logging.info(f"All columns of each excel file will be loaded.")

# <-1->    <setting_excel> 'usecols' CORRECT INPUT MESSAGE
            elif type(setting_excel.get('usecols')) in [list, range, int, str]:
                logging.info(f"The columns of each excel file to load are specified as: {setting_excel['usecols']}")

# <-1->    <setting_excel> 'usecols' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'usecols': {5}}   =>   setting_excel = {'usecols': None}
            else:
                raise Exception
        except:
            setting_excel['usecols'] = None
            logging.warning(f"Incorrectly defined <setting_excel> usecols variable.")
            logging.warning(f"The <setting_excel> usecols has been set to the following default value: '{setting_excel['usecols']}'")
            logging.warning(f"All columns of each excel file will be loaded.")
            warnings = True

# <-1->    <setting_excel> 'displayed_decimals' STRING TO INTEGER CONVERSION
#          setting_excel = {'displayed_decimals': '5'}   =>   setting_excel = {'displayed_decimals': 5}
        try:
            if type(setting_excel['displayed_decimals']) == str:
                setting_excel['displayed_decimals'] = int(setting_excel['displayed_decimals'])
#          setting_excel = {'displayed_decimals': 'one'}   <=>   setting_excel = {'displayed_decimals': 'one'}        
        except:
            pass

# <-1->    <setting_excel> 'displayed_decimals' DEFAULT VALUE ASSIGNMENT
#          setting_excel = {}   =>   setting_excel = {'displayed_decimals': 4}
        default_decimals = 4
        if setting_excel.get('displayed_decimals') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_excel['displayed_decimals'] = default_decimals
            logging.info(f"The <setting_excel> displayed_decimals has been set to the following default value: {setting_excel['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_excel['displayed_decimals']}")

# <-1->    <setting_excel> 'displayed_decimals' CORRECT INPUT MESSAGE
        elif type(setting_excel.get('displayed_decimals')) == int:
            logging.info(f"The <setting_excel> displayed_decimals is set to: {setting_excel['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_excel['displayed_decimals']}")

# <-1->    <setting_excel> 'displayed_decimals' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'displayed_decimals': 'one'}   =>   setting_excel = {'displayed_decimals': 4}
        elif type(setting_excel.get('displayed_decimals')) != int:
            setting_excel['displayed_decimals'] = default_decimals
            logging.warning(f"The <setting_excel> displayed_decimals value must be an integer.")
            logging.warning(f"The <setting_excel> displayed_decimals has been set to the following default value: {setting_excel['displayed_decimals']}")
            logging.warning(f"The number of displayed floating point decimals is: {setting_excel['displayed_decimals']}")
            warnings = True
    
# <-1->    <setting_excel> 'sheet_name' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_excel = {}   =>   setting_excel = {'sheet_name': 0}
        if setting_excel.get('sheet_name') in [None, [], (), {}, '']:
            setting_excel['sheet_name'] = 0
            logging.warning(f"The <setting_excel> sheet_name has not been defined.")
            logging.warning(f"The <setting_excel> sheet_name has been set to the following default value: {setting_excel['sheet_name']}")
            logging.warning(f"Only the first sheet of each excel file will be loaded.")
            warnings = True

# <-1->    <setting_excel> 'sheet_name' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'sheet_name': True}   =>   setting_excel = {'sheet_name': 0}
        elif type(setting_excel.get('sheet_name')) == bool:
            setting_excel['sheet_name'] = 0 
            logging.warning(f"Incorrectly defined <setting_excel> sheet_name variable.")
            logging.warning(f"The <setting_excel> sheet_name has been set to the following default value: {setting_excel['sheet_name']}")
            logging.warning(f"Only the first sheet of each excel file will be loaded.")
            warnings = True

# <-1->    <setting_excel> 'sheet_name' TUPLE TO LIST CONVERSION
#          setting_excel = {'sheet_name': (1,2,3)}   =>   setting_excel = {'sheet_name': [1, 2, 3]}
        elif type(setting_excel.get('sheet_name')) == tuple:
            setting_excel['sheet_name'] = list(setting_excel['sheet_name'])
            logging.info(f"The sheet names/indexes to load are specified as: {setting_excel['sheet_name']}")

# <-1->    <setting_excel> 'sheet_name' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_excel = {'sheet_name': {10}}   =>   setting_excel = {'sheet_name': 0}
        elif not isinstance(setting_excel.get('sheet_name'), (str, int, list)):
            setting_excel['sheet_name'] = 0 
            logging.warning(f"Incorrectly defined <setting_excel> sheet_name variable.")
            logging.warning(f"The <setting_excel> sheet_name has been set to the following default value: {setting_excel['sheet_name']}")
            logging.warning(f"Only the first sheet of each excel file will be loaded.")
            warnings = True
        
# <-1->    <setting_excel> 'sheet_name'  CORRECT INPUT MESSAGE
        else:
            logging.info(f"The sheet names/indexes to load are specified as: {setting_excel['sheet_name']}")
            if setting_excel.get('sheet_name') == 0:
                logging.info(f"Only the first sheet of each excel file will be loaded.")
            
# <-1->    <setting_excel> DECIMAL SEPARATOR AND SETTING DISPLAY MESSAGE
        logging.info(f'The excel read setting is set as follows: {list(setting_excel.items())}')
        logging.info(f'The decimal separator will be automatically detected.')

#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    TABLE DATA READ FUNCTION INPUT CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
        
# <-0->    <setting_table> VARIABLE SETTING SECTION
    elif data_read_option == ['pd.read_table']:

# <-1->    DICTIONARY UNPACKING FUNCTION DEFINITION
#          item = [[(([{'displayed_decimals': 5}]))]]   =>   dict_unpack(item)   =>   item = {'displayed_decimals': 5}
        def dict_unpack(item):
            if isinstance(item, list):
                return dict_unpack(item[0])
            elif isinstance(item, dict):
                return item
        
# <-1->    <setting_table> VARIABLE UNPACKING
#          setting_table = [[(([{'dtype': 'np.float64'}]))]]   =>   setting_table = {'dtype': 'np.float64'}
        setting_table = dict_unpack(setting_table)

# <-1->    INCORRECT <setting_table> VARIABLE TYPE
#          setting_table = [{'displayed_decimals': 5, 'dtype': 'np.float64', 'encoding': 'utf-8'}]
        if type(setting_table) != dict:
            raise Exception(f"\nIncorrectly defined <setting_table> variable.\nThe <setting_table> variable must be a dictionary.")

# <-1->    DICTIONARY ITEM UNPACKING FUNCTION DEFINITION
#          item = {'dict_key': ([(['value'])])}   =>   unpack_list(item)   =>   item = {'dict_key': 'value'}
        def unpack_list(item):
            if isinstance(item, list):
                return unpack_list(item[-1])
            else:
                return item

# <-1->    <setting_table> VARIABLE DICTIONARY VALUE LIST UNPACKING
#          setting_table = {'displayed_decimals': ([([5])])}   =>   setting_table = {'displayed_decimals': 5}
        items_to_unpack = ['displayed_decimals', 'dtype', 'encoding', 'skiprows', 'skipfooter', 'delimiter', 'engine']
        for key, value in setting_table.items():
            if key.lower() in items_to_unpack:
                setting_table[key] = unpack_list(value)

# <-1->    DUMMY DICTIONARY AND ACCEPTED DICTIONARY KEY ITEMS VARIABLE CREATION
        setting_table_accepted_keys = ['displayed_decimals', 'dtype', 'encoding', 'skiprows', 'skipfooter', 'usecols', 'delimiter', 'engine']
        unknown_keys = {}

# <-1->    <setting_table> VARIABLE LOWER CASE CONVERTION
#          setting_table = {'dispLayed_dEciMals': 5}   =>   setting_table = {'displayed_decimals': 5}
        setting_table = {key.lower(): value.lower() if isinstance(value, str) and key.lower() not in ['usecols'] else value for key, value in setting_table.items()}

# <-1->    DICTIONARY DUPLICATE FILTRATION FUNCTION DEFINITION
#          input_dict = {'skiprows': 0, 'skiprows': 0}   =>   input_dict = {'skiprows': 0}
        def remove_duplicate_dict(input_dict):
            dict_result = {}
            for key, value in input_dict.items():
                dict_result.setdefault(key, value)
            return dict_result

# <-1->    <setting_table> VARIABLE DUPLICATE FILTRATION
#          setting_table = {'displayed_decimals': '0', 'displayed_decimals': '0'}   =>   setting_table = {'displayed_decimals': '0'}
        setting_table = remove_duplicate_dict(setting_table)

# <-1->    <setting_table> VARIABLE UNKNOWN KEY SEARCH
#          setting_table = {'skipipipirows': '0', 'skiprows': '0'} AND setting_table_accepted_keys = ['skiprows']   =>   unknown_keys = {'skipipipirows': 0'}
        for key, value in setting_table.items():
            if key.lower() not in setting_table_accepted_keys:
                unknown_keys.setdefault(key, value)

# <-1->    <setting_table> VARIABLE UNKNOWN ITEM FILTRATION
        if unknown_keys != {}:
            logging.warning(f"The following <setting_table> inputs have not been recognized: {unknown_keys}")
#          setting_table = {'skiprows': '0', 'astype': 'np.float64'} AND unknown_keys = {'astype': 'np.float64'}   =>   setting_table = {'skiprows': '0'}
            setting_table = {key: value for key, value in setting_table.items() if key not in unknown_keys.keys()}
            warnings = True
        elif unknown_keys == {} and len(setting_table) > 0:
            logging.info(f"All <setting_table> inputs have been recognized.")
    
# <-1->    <setting_table> VARIABLE UNDEFINED ITEMS NONE VALUE ASSIGNMENT
#          setting_table = {'dtype': 'float64'}  AND  setting_table_accepted_keys = ['engine', 'dtype']   =>   setting_table = {'dtype': 'float64', 'engine': None}
        for item in setting_table_accepted_keys:
            if item not in setting_table.keys():
                setting_table.setdefault(item, None)
        
# <-1->    <setting_table> 'dtype' DEFAULT VALUE CREATION
        default_dtype = np.float64

# <-1->    <setting_table> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_table = {'dtype': None}   =>   setting_table = {'dtype': 'float64'}
        if setting_table.get('dtype') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_table['dtype'] = default_dtype
            logging.warning(f"The <setting_table> dtype has not been defined.")
            logging.warning(f"The <setting_table> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_table> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_table={} => setting_table={'dtype':'np.float64'}  | setting_table={'dtype':'np.float16'} <=> setting_table={'dtype':'np.float16'}
        add_setting_table = {'dtype': default_dtype}
        [setting_table.setdefault(key, value) for key, value in add_setting_table.items()]

# <-1->    <setting_table> 'dtype' KEY INPUT STRING MODIFICATION
#          setting_table = {'dtype':'np.float64'}   =>   setting_table = {'dtype':'float64'}
        if type(setting_table['dtype']) == str:
            setting_table['dtype'] = setting_table['dtype'].replace('np.','')

# <-1->    <setting_table> 'dtype' KEY NUMPY VALUE ASSIGNMENT
#          setting_table = {'dtype': 'float64'}   =>   setting_table = {'dtype': np.float64}
            try:
                setting_table['dtype'] = getattr(np, setting_table['dtype'])
#          setting_table = {'dtype': 'float1000'}   =>   setting_table = {'dtype': None}
            except:
                setting_table['dtype'] = None

# <-1->    <setting_table> 'dtype' KEY INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'dtype': None}   =>   setting_table = {'dtype': 'float64'}
        if setting_table['dtype'] not in [np.float64, np.float32, np.float16, np.int64, np.int32, np.int16,\
                                    np.uint64, np.uint32, np.uint16, int, float]:
            setting_table['dtype'] = default_dtype
            logging.warning(f"Incorrectly defined <setting_table> dtype.")
            logging.warning(f"The <setting_table> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_table> 'dtype' KEY ASSIGNED VALUE DISPLAY
        logging.info(f"The <setting_table> dtype is set to: '{setting_table['dtype']}'")

# <-1->    <accepted_engine> VARIABLE CREATION
        accepted_engine = ['python', 'c', 'pyarrow']

# <-1->    <setting_table> 'engine' DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'engine': 'python'}
        if setting_table.get('engine') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_table['engine'] = 'python'
            logging.info(f"The <setting_table> engine has been set to the following default value: '{setting_table['engine']}'")

# <-1->    <setting_table> 'engine' CORRECT INPUT MESSAGE
        elif setting_table.get('engine') in accepted_engine:
            logging.info(f"The <setting_table> engine is set to: '{setting_table['engine']}'")

# <-1->    <setting_table> 'engine' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'engine': 'ABC'}   =>   setting_table = {'engine': 'python'}
        elif setting_table.get('engine') not in accepted_engine:
            setting_table['engine'] = 'python'
            logging.info(f"The accepted <setting_table> engine values are: {accepted_engine}")
            logging.warning(f"Incorrectly defined <setting_table> engine.")
            logging.warning(f"The <setting_table> engine has been set to the following default value: '{setting_table['engine']}'")
            warnings = True
    
# <-1->    <setting_table> 'skiprows' STRING TO INTEGER CONVERSION
#          setting_table = {'skiprows': '0'}   =>   setting_table = {'skiprows': 0}
        try:
            if type(setting_table['skiprows']) == str:
                setting_table['skiprows'] = int(setting_table['skiprows'])
#          setting_table = {'skiprows': 'one'}   <=>   setting_table = {'skiprows': 'one'}
        except:
            pass

# <-1->    <setting_table> 'skiprows' DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'skiprows': 0}
        if setting_table.get('skiprows') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_table['skiprows'] = 0
            logging.info(f"The <setting_table> skiprows has been set to the following default value: {setting_table['skiprows']}")
            logging.info(f"No initial rows have been skipped in the data loading process.")

# <-1->    <setting_table> 'skiprows' CORRECT INPUT MESSAGE
        elif type(setting_table.get('skiprows')) == int:
            logging.info(f"The <setting_table> skiprows is set to: {setting_table['skiprows']}")
            logging.info(f"The number of skipped initial rows in the data loading process was: {setting_table['skiprows']}")

# <-1->    <setting_table> 'skiprows' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'skiprows': 'one'}   =>   setting_table = {'skiprows': 0}
        elif type(setting_table.get('skiprows')) != int:
            setting_table['skiprows'] = 0
            logging.warning(f"The <setting_table> skiprows value must be an integer.")
            logging.warning(f"The <setting_table> skiprows has been set to the following default value: {setting_table['skiprows']}")
            logging.warning(f"No initial rows have been skipped in the data loading process.")
            warnings = True
    
# <-1->    <setting_table> 'skipfooter' STRING TO INTEGER CONVERSION
#          setting_table = {'skipfooter': '0'}   =>   setting_table = {'skipfooter': 0}
        try:
            if type(setting_table['skipfooter']) == str:
                setting_table['skipfooter'] = int(setting_table['skipfooter'])
#          setting_table = {'skipfooter': 'one'}   <=>   setting_table = {'skipfooter': 'one'}
        except:
            pass

# <-1->    <setting_table> 'skipfooter' DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'skipfooter': 0}
        if setting_table.get('skipfooter') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_table['skipfooter'] = 0
            logging.info(f"The <setting_table> skipfooter has been set to the following default value: {setting_table['skipfooter']}")
            logging.info(f"No end rows have been skipped in the data loading process.")

# <-1->    <setting_table> 'skipfooter' CORRECT INPUT MESSAGE
        elif type(setting_table.get('skipfooter')) == int:
            logging.info(f"The <setting_table> skipfooter is set to: {setting_table['skipfooter']}")
            logging.info(f"The number of skipped end rows in the data loading process was: {setting_table['skipfooter']}")

# <-1->    <setting_table> 'skipfooter' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'skipfooter': 'one'}   =>   setting_table = {'skipfooter': 0}
        elif type(setting_table.get('skipfooter')) != int:
            setting_table['skipfooter'] = 0
            logging.warning(f"The <setting_table> skipfooter value must be an integer.")
            logging.warning(f"The <setting_table> skipfooter has been set to the following default value: {setting_table['skipfooter']}")
            logging.warning(f"No end rows have been skipped in the data loading process.")
            warnings = True
    
# <-1->    <setting_table> 'usecols' TUPLE TO LIST CONVERSION
#          setting_table = {'usecols': (1,2,3)}   =>   setting_table = {'usecols': [1, 2, 3]}
        try:
            if type(setting_table.get('usecols')) in [tuple]:
                setting_table['usecols'] = list(setting_table['usecols'])
            
# <-1->    <setting_table> 'usecols' STRING OR INTEGER TO LIST CONVERSION
#          setting_table = {'usecols': 1}   =>   setting_table = {'usecols': [1]}
            elif type(setting_table.get('usecols')) in [str, int]:
                setting_table['usecols'] = [setting_table['usecols']]
            
# <-1->    <setting_table> 'usecols' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'usecols': None}
            if setting_table.get('usecols') in [[''], [' '], [], (), {}, None, '', ' ']:
                if setting_table.get('usecols') in [None]:
                    logging.info(f"The columns of each data file to load are specified as: {setting_table['usecols']}")
                else:
                    setting_table['usecols'] = None
                    logging.info(f"The <setting_table> usecols has been set to the following default value: '{setting_table['usecols']}'")
                logging.info(f"All columns of each data file will be loaded.")
            
# <-1->    <setting_table> 'usecols' CORRECT INPUT MESSAGE
            elif type(setting_table.get('usecols')) in [list, range, int, str]:
                logging.info(f"The columns of each data file to load are specified as: {setting_table['usecols']}")
    
# <-1->    <setting_table> 'usecols' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'usecols': {5}}   =>   setting_table = {'usecols': None}
            else:
                raise Exception
        except:
            setting_table['usecols'] = None
            logging.warning(f"Incorrectly defined <setting_table> usecols variable.")
            logging.warning(f"The <setting_table> usecols has been set to the following default value: '{setting_table['usecols']}'")
            logging.warning(f"All columns of each data file will be loaded.")
            warnings = True   
    
# <-1->    <setting_table> 'displayed_decimals' STRING TO INTEGER CONVERSION
#          setting_table = {'displayed_decimals': '5'}   =>   setting_table = {'displayed_decimals': 5}
        try:
            if type(setting_table['displayed_decimals']) == str:
                setting_table['displayed_decimals'] = int(setting_table['displayed_decimals'])
#          setting_table = {'displayed_decimals': 'one'}   <=>   setting_table = {'displayed_decimals': 'one'}        
        except:
            pass

# <-1->    <setting_table> 'displayed_decimals' DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'displayed_decimals': 4}
        default_decimals = 4
        if setting_table.get('displayed_decimals') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_table['displayed_decimals'] = default_decimals
            logging.info(f"The <setting_table> displayed_decimals has been set to the following default value: {setting_table['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_table['displayed_decimals']}")

# <-1->    <setting_table> 'displayed_decimals' CORRECT INPUT MESSAGE
        elif type(setting_table.get('displayed_decimals')) == int:
            logging.info(f"The <setting_table> displayed_decimals is set to: {setting_table['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_table['displayed_decimals']}")

# <-1->    <setting_table> 'displayed_decimals' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'displayed_decimals': 'one'}   =>   setting_table = {'displayed_decimals': 4}
        elif type(setting_table.get('displayed_decimals')) != int:
            setting_table['displayed_decimals'] = default_decimals
            logging.warning(f"The <setting_table> displayed_decimals value must be an integer.")
            logging.warning(f"The <setting_table> displayed_decimals has been set to the following default value: {setting_table['displayed_decimals']}")
            logging.warning(f"The number of displayed floating point decimals is: {setting_table['displayed_decimals']}")
            warnings = True
    
# <-1->    <accepted_encoding> VARIABLE CREATION
        accepted_encoding = ['utf-8', 'utf-16', 'utf-32', 'ascii']

# <-1->    <setting_table> 'encoding' DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'encoding': 'utf-8'}
        if setting_table.get('encoding') in [[''], [' '], [], (), {}, None, '', ' ']:
            add_setting_table = {'encoding':'utf-8'}
            setting_table['encoding'] = 'utf-8'
            logging.info(f"The <setting_table> encoding has been set to the following default value: '{setting_table['encoding']}'")

# <-1->    <setting_table> 'encoding' CORRECT INPUT MESSAGE
        elif setting_table.get('encoding') in accepted_encoding:
            logging.info(f"The <setting_table> encoding is set to: '{setting_table['encoding']}'")

# <-1->    <setting_table> 'encoding' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'encoding': ''}   =>   setting_table = {'encoding': 'utf-8'}
        elif setting_table.get('encoding') not in accepted_encoding:
            setting_table['encoding'] = 'utf-8'
            logging.info(f"The accepted <setting_table> encoding values are: {accepted_encoding}")
            logging.warning(f"Incorrectly defined <setting_table> encoding.")
            logging.warning(f"The <setting_table> encoding has been set to the following default value: '{setting_table['encoding']}'")
            warnings = True
    
# <-1->    <accepted_delimiter> VARIABLE CREATION
        accepted_delimiter = ["\t", " ", ":", "_", "-", "/", "|", "#", "\n", ";", ".", ",", "/", "\\", None, "default", r"\s"]

# <-1->    <setting_table> 'delimiter' DEFAULT VALUE ASSIGNMENT
#          setting_table = {}   =>   setting_table = {'delimiter': None}
        if setting_table.get('delimiter') in [None, "None", "default", (), [], {}, [''], [' ']]:
            add_setting_table = {'delimiter': None}
            setting_table['delimiter'] = None
            logging.info(f"The <setting_table> delimiter has been set to the following default value: '{setting_table['delimiter']}'")
            logging.info(f"The automatic delimiter detection option has been selected.")

# <-1->    <setting_table> 'delimiter' CORRECT INPUT MESSAGE
        elif setting_table.get('delimiter') in accepted_delimiter:
            logging.info(f"The <setting_table> delimiter is set to: {repr(setting_table['delimiter'])}")
        
# <-1->    <setting_table> 'delimiter' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_table = {'delimiter': ''}   =>   setting_table = {'delimiter': 'None'}
        elif setting_table.get('delimiter') not in accepted_delimiter:
            setting_table['delimiter'] = None
            logging.info(f"The accepted <setting_table> delimiter values are: {accepted_delimiter}")
            logging.warning(f"Incorrectly defined <setting_table> delimiter.")
            logging.warning(f"The <setting_table> delimiter has been set to the following default value: '{setting_table['delimiter']}'")
            logging.info(f"The automatic delimiter detection option has been selected.")
            warnings = True
            
# <-1->    <setting_table> SETTING DISPLAY MESSAGE
        logging.info(f'The table read setting is set as follows: {list(setting_table.items())}')
    
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    CSV DATA READ FUNCTION INPUT CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°  
    
# <-0->    <setting_csv> VARIABLE SETTING SECTION
    elif data_read_option == ['pd.read_csv']:

# <-1->    DICTIONARY UNPACKING FUNCTION DEFINITION
#          item = [[(([{'displayed_decimals': 5}]))]]   =>   dict_unpack(item)   =>   item = {'displayed_decimals': 5}
        def dict_unpack(item):
            if isinstance(item, list):
                return dict_unpack(item[0])
            elif isinstance(item, dict):
                return item
        
# <-1->    <setting_csv> VARIABLE UNPACKING
#          setting_csv = [[(([{'dtype': 'np.float64'}]))]]   =>   setting_csv = {'dtype': 'np.float64'}
        setting_csv = dict_unpack(setting_csv)

# <-1->    INCORRECT <setting_csv> VARIABLE TYPE
#          setting_csv = [{'displayed_decimals': 5, 'dtype': 'np.float64', 'encoding': 'utf-8'}]
        if type(setting_csv) != dict:
            raise Exception(f"\nIncorrectly defined <setting_csv> variable.\nThe <setting_csv> variable must be a dictionary.")

# <-1->    DICTIONARY ITEM UNPACKING FUNCTION DEFINITION
#          item = {'dict_key': ([(['value'])])}   =>   unpack_list(item)   =>   item = {'dict_key': 'value'}
        def unpack_list(item):
            if isinstance(item, list):
                return unpack_list(item[-1])
            else:
                return item

# <-1->    <setting_csv> VARIABLE DICTIONARY VALUE LIST UNPACKING
#          setting_csv = {'displayed_decimals': ([([5])])}   =>   setting_csv = {'displayed_decimals': 5}
        items_to_unpack = ['displayed_decimals', 'dtype', 'encoding', 'skiprows', 'skipfooter', 'delimiter', 'engine']
        for key, value in setting_csv.items():
            if key.lower() in items_to_unpack:
                setting_csv[key] = unpack_list(value)

# <-1->    DUMMY DICTIONARY AND ACCEPTED DICTIONARY KEY ITEMS VARIABLE CREATION
        setting_csv_accepted_keys = ['displayed_decimals', 'dtype', 'encoding', 'skiprows', 'skipfooter', 'usecols', 'delimiter', 'engine']
        unknown_keys = {}

# <-1->    <setting_csv> VARIABLE LOWER CASE CONVERTION
#          setting_csv = {'dispLayed_dEciMals': 5}   =>   setting_csv = {'displayed_decimals': 5}
        setting_csv = {key.lower(): value.lower() if isinstance(value, str) and key.lower() not in ['usecols'] else value for key, value in setting_csv.items()}

# <-1->    DICTIONARY DUPLICATE FILTRATION FUNCTION DEFINITION
#          input_dict = {'skiprows': 0, 'skiprows': 0}   =>   input_dict = {'skiprows': 0}
        def remove_duplicate_dict(input_dict):
            dict_result = {}
            for key, value in input_dict.items():
                dict_result.setdefault(key, value)
            return dict_result

# <-1->    <setting_csv> VARIABLE DUPLICATE FILTRATION
#          setting_csv = {'displayed_decimals': '0', 'displayed_decimals': '0'}   =>   setting_csv = {'displayed_decimals': '0'}
        setting_csv = remove_duplicate_dict(setting_csv)

# <-1->    <setting_csv> VARIABLE UNKNOWN KEY SEARCH
#          setting_csv = {'skipipipirows': '0', 'skiprows': '0'} AND setting_csv_accepted_keys = ['skiprows']   =>   unknown_keys = {'skipipipirows': 0'}
        for key, value in setting_csv.items():
            if key.lower() not in setting_csv_accepted_keys:
                unknown_keys.setdefault(key, value)

# <-1->    <setting_csv> VARIABLE UNKNOWN ITEM FILTRATION
        if unknown_keys != {}:
            logging.warning(f"The following <setting_csv> inputs have not been recognized: {unknown_keys}")
#          setting_csv = {'skiprows': '0', 'astype': 'np.float64'} AND unknown_keys = {'astype': 'np.float64'}   =>   setting_csv = {'skiprows': '0'}
            setting_csv = {key: value for key, value in setting_csv.items() if key not in unknown_keys.keys()}
            warnings = True
        elif unknown_keys == {} and len(setting_csv) > 0:
            logging.info(f"All <setting_csv> inputs have been recognized.")
    
# <-1->    <setting_csv> VARIABLE UNDEFINED ITEMS NONE VALUE ASSIGNMENT
#          setting_csv = {'dtype': 'float64'}  AND  setting_csv_accepted_keys = ['engine', 'dtype']   =>   setting_csv = {'dtype': 'float64', 'engine': None}
        for item in setting_csv_accepted_keys:
            if item not in setting_csv.keys():
                setting_csv.setdefault(item, None)
        
# <-1->    <setting_csv> 'dtype' DEFAULT VALUE CREATION
        default_dtype = np.float64

# <-1->    <setting_csv> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'dtype': None}   =>   setting_csv = {'dtype': 'float64'}
        if setting_csv.get('dtype') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_csv['dtype'] = default_dtype
            logging.warning(f"The <setting_csv> dtype has not been defined.")
            logging.warning(f"The <setting_csv> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_csv> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_csv = {} => setting_csv = {'dtype':'np.float64'}  | setting_csv = {'dtype':'np.float16'} <=> setting_csv = {'dtype':'np.float16'}
        add_setting_csv = {'dtype': default_dtype}
        [setting_csv.setdefault(key, value) for key, value in add_setting_csv.items()]

# <-1->    <setting_csv> 'dtype' KEY INPUT STRING MODIFICATION
#          setting_csv = {'dtype':'np.float64'}   =>   setting_csv = {'dtype':'float64'}
        if type(setting_csv['dtype']) == str:
            setting_csv['dtype'] = setting_csv['dtype'].replace('np.','')

# <-1->    <setting_csv> 'dtype' KEY NUMPY VALUE ASSIGNMENT
#          setting_csv = {'dtype': 'float64'}   =>   setting_csv = {'dtype': np.float64}
            try:
                setting_csv['dtype'] = getattr(np, setting_csv['dtype'])
#          setting_csv = {'dtype': 'float1000'}   =>   setting_csv = {'dtype': None}
            except:
                setting_csv['dtype'] = None

# <-1->    <setting_csv> 'dtype' KEY INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'dtype': None}   =>   setting_csv = {'dtype': 'float64'}
        if setting_csv['dtype'] not in [np.float64, np.float32, np.float16, np.int64, np.int32, np.int16,\
                                    np.uint64, np.uint32, np.uint16, int, float]:
            setting_csv['dtype'] = default_dtype
            logging.warning(f"Incorrectly defined <setting_csv> dtype.")
            logging.warning(f"The <setting_csv> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_csv> 'dtype' KEY ASSIGNED VALUE DISPLAY
        logging.info(f"The <setting_csv> dtype is set to: '{setting_csv['dtype']}'")

# <-1->    <accepted_engine> VARIABLE CREATION
        accepted_engine = ['python', 'c', 'pyarrow']

# <-1->    <setting_csv> 'engine' DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'engine': 'python'}
        if setting_csv.get('engine') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_csv['engine'] = 'python'
            logging.info(f"The <setting_csv> engine has been set to the following default value: '{setting_csv['engine']}'")

# <-1->    <setting_csv> 'engine' CORRECT INPUT MESSAGE
        elif setting_csv.get('engine') in accepted_engine:
            logging.info(f"The <setting_csv> engine is set to: '{setting_csv['engine']}'")

# <-1->    <setting_csv> 'engine' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'engine': 'ABC'}   =>   setting_csv = {'engine': 'python'}
        elif setting_csv.get('engine') not in accepted_engine:
            setting_csv['engine'] = 'python'
            logging.info(f"The accepted <setting_csv> engine values are: {accepted_engine}")
            logging.warning(f"Incorrectly defined <setting_csv> engine.")
            logging.warning(f"The <setting_csv> engine has been set to the following default value: '{setting_csv['engine']}'")
            warnings = True
    
# <-1->    <setting_csv> 'skiprows' STRING TO INTEGER CONVERSION
#          setting_csv = {'skiprows': '0'}   =>   setting_csv = {'skiprows': 0}
        try:
            if type(setting_csv['skiprows']) == str:
                setting_csv['skiprows'] = int(setting_csv['skiprows'])
#          setting_csv = {'skiprows': 'one'}   <=>   setting_csv = {'skiprows': 'one'}
        except:
            pass

# <-1->    <setting_csv> 'skiprows' DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'skiprows': 0}
        if setting_csv.get('skiprows') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_csv['skiprows'] = 0
            logging.info(f"The <setting_csv> skiprows has been set to the following default value: {setting_csv['skiprows']}")
            logging.info(f"No initial rows have been skipped in the data loading process.")

# <-1->    <setting_csv> 'skiprows' CORRECT INPUT MESSAGE
        elif type(setting_csv.get('skiprows')) == int:
            logging.info(f"The <setting_csv> skiprows is set to: {setting_csv['skiprows']}")
            logging.info(f"The number of skipped initial rows in the data loading process was: {setting_csv['skiprows']}")

# <-1->    <setting_csv> 'skiprows' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'skiprows': 'one'}   =>   setting_csv = {'skiprows': 0}
        elif type(setting_csv.get('skiprows')) != int:
            setting_csv['skiprows'] = 0
            logging.warning(f"The <setting_csv> skiprows value must be an integer.")
            logging.warning(f"The <setting_csv> skiprows has been set to the following default value: {setting_csv['skiprows']}")
            logging.warning(f"No initial rows have been skipped in the data loading process.")
            warnings = True
    
# <-1->    <setting_csv> 'skipfooter' STRING TO INTEGER CONVERSION
#          setting_csv = {'skipfooter': '0'}   =>   setting_csv = {'skipfooter': 0}
        try:
            if type(setting_csv['skipfooter']) == str:
                setting_csv['skipfooter'] = int(setting_csv['skipfooter'])
#          setting_csv = {'skipfooter': 'one'}   <=>   setting_csv = {'skipfooter': 'one'}
        except:
            pass

# <-1->    <setting_csv> 'skipfooter' DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'skipfooter': 0}
        if setting_csv.get('skipfooter') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_csv['skipfooter'] = 0
            logging.info(f"The <setting_csv> skipfooter has been set to the following default value: {setting_csv['skipfooter']}")
            logging.info(f"No end rows have been skipped in the data loading process.")

# <-1->    <setting_csv> 'skipfooter' CORRECT INPUT MESSAGE
        elif type(setting_csv.get('skipfooter')) == int:
            logging.info(f"The <setting_csv> skipfooter is set to: {setting_csv['skipfooter']}")
            logging.info(f"The number of skipped end rows in the data loading process was: {setting_csv['skipfooter']}")

# <-1->    <setting_csv> 'skipfooter' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'skipfooter': 'one'}   =>   setting_csv = {'skipfooter': 0}
        elif type(setting_csv.get('skipfooter')) != int:
            setting_csv['skipfooter'] = 0
            logging.warning(f"The <setting_csv> skipfooter value must be an integer.")
            logging.warning(f"The <setting_csv> skipfooter has been set to the following default value: {setting_csv['skipfooter']}")
            logging.warning(f"No end rows have been skipped in the data loading process.")
            warnings = True
    
# <-1->    <setting_csv> 'usecols' TUPLE TO LIST CONVERSION
#          setting_csv = {'usecols': (1,2,3)}   =>   setting_csv = {'usecols': [1, 2, 3]}
        try:
            if type(setting_csv.get('usecols')) in [tuple]:
                setting_csv['usecols'] = list(setting_csv['usecols'])
            
# <-1->    <setting_csv> 'usecols' STRING OR INTEGER TO LIST CONVERSION
#          setting_csv = {'usecols': 1}   =>   setting_csv = {'usecols': [1]}
            elif type(setting_csv.get('usecols')) in [str, int]:
                setting_csv['usecols'] = [setting_csv['usecols']]
            
# <-1->    <setting_csv> 'usecols' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'usecols': None}
            if setting_csv.get('usecols') in [[''], [' '], [], (), {}, None, '', ' ']:
                if setting_csv.get('usecols') in [None]:
                    logging.info(f"The columns of each data file to load are specified as: {setting_csv['usecols']}")
                else:
                    setting_csv['usecols'] = None
                    logging.info(f"The <setting_csv> usecols has been set to the following default value: '{setting_csv['usecols']}'")
                logging.info(f"All columns of each data file will be loaded.")
            
# <-1->    <setting_csv> 'usecols' CORRECT INPUT MESSAGE
            elif type(setting_csv.get('usecols')) in [list, range, int, str]:
                logging.info(f"The columns of each data file to load are specified as: {setting_csv['usecols']}")
    
# <-1->    <setting_csv> 'usecols' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'usecols': {5}}   =>   setting_csv = {'usecols': None}
            else:
                raise Exception
        except:
            setting_csv['usecols'] = None
            logging.warning(f"Incorrectly defined <setting_csv> usecols variable.")
            logging.warning(f"The <setting_csv> usecols has been set to the following default value: '{setting_csv['usecols']}'")
            logging.warning(f"All columns of each data file will be loaded.")
            warnings = True   
    
# <-1->    <setting_csv> 'displayed_decimals' STRING TO INTEGER CONVERSION
#          setting_csv = {'displayed_decimals': '5'}   =>   setting_csv = {'displayed_decimals': 5}
        try:
            if type(setting_csv['displayed_decimals']) == str:
                setting_csv['displayed_decimals'] = int(setting_csv['displayed_decimals'])
#          setting_csv = {'displayed_decimals': 'one'}   <=>   setting_csv = {'displayed_decimals': 'one'}        
        except:
            pass

# <-1->    <setting_csv> 'displayed_decimals' DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'displayed_decimals': 4}
        default_decimals = 4
        if setting_csv.get('displayed_decimals') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_csv['displayed_decimals'] = default_decimals
            logging.info(f"The <setting_csv> displayed_decimals has been set to the following default value: {setting_csv['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_csv['displayed_decimals']}")

# <-1->    <setting_csv> 'displayed_decimals' CORRECT INPUT MESSAGE
        elif type(setting_csv.get('displayed_decimals')) == int:
            logging.info(f"The <setting_csv> displayed_decimals is set to: {setting_csv['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_csv['displayed_decimals']}")

# <-1->    <setting_csv> 'displayed_decimals' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'displayed_decimals': 'one'}   =>   setting_csv = {'displayed_decimals': 4}
        elif type(setting_csv.get('displayed_decimals')) != int:
            setting_csv['displayed_decimals'] = default_decimals
            logging.warning(f"The <setting_csv> displayed_decimals value must be an integer.")
            logging.warning(f"The <setting_csv> displayed_decimals has been set to the following default value: {setting_csv['displayed_decimals']}")
            logging.warning(f"The number of displayed floating point decimals is: {setting_csv['displayed_decimals']}")
            warnings = True
    
# <-1->    <accepted_encoding> VARIABLE CREATION
        accepted_encoding = ['utf-8', 'utf-16', 'utf-32', 'ascii']

# <-1->    <setting_csv> 'encoding' DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'encoding': 'utf-8'}
        if setting_csv.get('encoding') in [[''], [' '], [], (), {}, None, '', ' ']:
            add_setting_csv = {'encoding':'utf-8'}
            setting_csv['encoding'] = 'utf-8'
            logging.info(f"The <setting_csv> encoding has been set to the following default value: '{setting_csv['encoding']}'")

# <-1->    <setting_csv> 'encoding' CORRECT INPUT MESSAGE
        elif setting_csv.get('encoding') in accepted_encoding:
            logging.info(f"The <setting_csv> encoding is set to: '{setting_csv['encoding']}'")

# <-1->    <setting_csv> 'encoding' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'encoding': ''}   =>   setting_csv = {'encoding': 'utf-8'}
        elif setting_csv.get('encoding') not in accepted_encoding:
            setting_csv['encoding'] = 'utf-8'
            logging.info(f"The accepted <setting_csv> encoding values are: {accepted_encoding}")
            logging.warning(f"Incorrectly defined <setting_csv> encoding.")
            logging.warning(f"The <setting_csv> encoding has been set to the following default value: '{setting_csv['encoding']}'")
            warnings = True
    
# <-1->    <accepted_delimiter> VARIABLE CREATION
        accepted_delimiter = ["\t", " ", ":", "_", "-", "/", "|", "#", "\n", ";", ".", ",", "/", "\\", None, "default", r"\s"]

# <-1->    <setting_csv> 'delimiter' DEFAULT VALUE ASSIGNMENT
#          setting_csv = {}   =>   setting_csv = {'delimiter': None}
        if setting_csv.get('delimiter') in [None, "None", "default", (), [], {}, [''], [' ']]:
            add_setting_csv = {'delimiter': None}
            setting_csv['delimiter'] = None
            logging.info(f"The <setting_csv> delimiter has been set to the following default value: '{setting_csv['delimiter']}'")
            logging.info(f"The automatic delimiter detection option has been selected.")

# <-1->    <setting_csv> 'delimiter' CORRECT INPUT MESSAGE
        elif setting_csv.get('delimiter') in accepted_delimiter:
            logging.info(f"The <setting_csv> delimiter is set to: {repr(setting_csv['delimiter'])}")
        
# <-1->    <setting_csv> 'delimiter' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_csv = {'delimiter': ''}   =>   setting_csv = {'delimiter': 'None'}
        elif setting_csv.get('delimiter') not in accepted_delimiter:
            setting_csv['delimiter'] = None
            logging.info(f"The accepted <setting_csv> delimiter values are: {accepted_delimiter}")
            logging.warning(f"Incorrectly defined <setting_csv> delimiter.")
            logging.warning(f"The <setting_csv> delimiter has been set to the following default value: '{setting_csv['delimiter']}'")
            logging.info(f"The automatic delimiter detection option has been selected.")
            warnings = True
            
# <-1->    <setting_csv> SETTING DISPLAY MESSAGE
        logging.info(f'The csv read setting is set as follows: {list(setting_csv.items())}')
    
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    FWF DATA READ FUNCTION INPUT CHECK    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°

# <-0->    <setting_fwf> VARIABLE SETTING SECTION
    elif data_read_option == ['pd.read_fwf']:

# <-1->    DICTIONARY UNPACKING FUNCTION DEFINITION
#          item = [[(([{'displayed_decimals': 5}]))]]   =>   dict_unpack(item)   =>   item = {'displayed_decimals': 5}
        def dict_unpack(item):
            if isinstance(item, list):
                return dict_unpack(item[0])
            elif isinstance(item, dict):
                return item
        
# <-1->    <setting_fwf> VARIABLE UNPACKING
#          setting_fwf = [[(([{'dtype': 'np.float64'}]))]]   =>   setting_fwf = {'dtype': 'np.float64'}
        setting_fwf = dict_unpack(setting_fwf)

# <-1->    INCORRECT <setting_fwf> VARIABLE TYPE
#          setting_fwf = [{'displayed_decimals': 5, 'dtype': 'np.float64', 'encoding': 'utf-8'}]
        if type(setting_fwf) != dict:
            raise Exception(f"\nIncorrectly defined <setting_fwf> variable.\nThe <setting_fwf> variable must be a dictionary.")

# <-1->    DICTIONARY ITEM UNPACKING FUNCTION DEFINITION
#          item = {'dict_key': ([(['value'])])}   =>   unpack_list(item)   =>   item = {'dict_key': 'value'}
        def unpack_list(item):
            if isinstance(item, list):
                return unpack_list(item[-1])
            else:
                return item

# <-1->    <setting_fwf> VARIABLE DICTIONARY VALUE LIST UNPACKING
#          setting_fwf = {'displayed_decimals': ([([5])])}   =>   setting_fwf = {'displayed_decimals': 5}
        items_to_unpack = ['displayed_decimals', 'dtype', 'encoding', 'skiprows', 'skipfooter', 'engine']
        for key, value in setting_fwf.items():
            if key.lower() in items_to_unpack:
                setting_fwf[key] = unpack_list(value)

# <-1->    DUMMY DICTIONARY AND ACCEPTED DICTIONARY KEY ITEMS VARIABLE CREATION
        setting_fwf_accepted_keys = ['displayed_decimals', 'dtype', 'encoding', 'skiprows', 'skipfooter', 'engine']
        unknown_keys = {}

# <-1->    <setting_fwf> VARIABLE LOWER CASE CONVERTION
#          setting_fwf = {'dispLayed_dEciMals': 5}   =>   setting_fwf = {'displayed_decimals': 5}
        setting_fwf = {key.lower(): value.lower() if isinstance(value, str) else value for key, value in setting_fwf.items()}
        
# <-1->    DICTIONARY DUPLICATE FILTRATION FUNCTION DEFINITION
#          input_dict = {'skiprows': 0, 'skiprows': 0}   =>   input_dict = {'skiprows': 0}
        def remove_duplicate_dict(input_dict):
            dict_result = {}
            for key, value in input_dict.items():
                dict_result.setdefault(key, value)
            return dict_result

# <-1->    <setting_fwf> VARIABLE DUPLICATE FILTRATION
#          setting_fwf = {'displayed_decimals': '0', 'displayed_decimals': '0'}   =>   setting_fwf = {'displayed_decimals': '0'}
        setting_fwf = remove_duplicate_dict(setting_fwf)

# <-1->    <setting_fwf> VARIABLE UNKNOWN KEY SEARCH
#          setting_fwf = {'skipipipirows': '0', 'skiprows': '0'} AND setting_fwf_accepted_keys = ['skiprows']   =>   unknown_keys = {'skipipipirows': 0'}
        for key, value in setting_fwf.items():
            if key.lower() not in setting_fwf_accepted_keys:
                unknown_keys.setdefault(key, value)

# <-1->    <setting_fwf> VARIABLE UNKNOWN ITEM FILTRATION
        if unknown_keys != {}:
            logging.warning(f"The following <setting_fwf> inputs have not been recognized: {unknown_keys}")
#          setting_fwf = {'skiprows': '0', 'astype': 'np.float64'} AND unknown_keys = {'astype': 'np.float64'}   =>   setting_fwf = {'skiprows': '0'}
            setting_fwf = {key: value for key, value in setting_fwf.items() if key not in unknown_keys.keys()}
            warnings = True
        elif unknown_keys == {} and len(setting_fwf) > 0:
            logging.info(f"All <setting_fwf> inputs have been recognized.")
    
# <-1->    <setting_fwf> VARIABLE UNDEFINED ITEMS NONE VALUE ASSIGNMENT
#          setting_fwf = {'dtype': 'float64'}  AND  setting_fwf_accepted_keys = ['engine', 'dtype']   =>   setting_fwf = {'dtype': 'float64', 'engine': None}
        for item in setting_fwf_accepted_keys:
            if item not in setting_fwf.keys():
                setting_fwf.setdefault(item, None)
        
# <-1->    <setting_fwf> 'dtype' DEFAULT VALUE CREATION
        default_dtype = np.float64

# <-1->    <setting_fwf> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'dtype': None}   =>   setting_fwf = {'dtype': 'float64'}
        if setting_fwf.get('dtype') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_fwf['dtype'] = default_dtype
            logging.warning(f"The <setting_fwf> dtype has not been defined.")
            logging.warning(f"The <setting_fwf> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_fwf> 'dtype' NOT DEFINED DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {} => setting_fwf = {'dtype':'np.float64'}  | setting_fwf = {'dtype':'np.float16'} <=> setting_fwf = {'dtype':'np.float16'}
        add_setting_fwf = {'dtype': default_dtype}
        [setting_fwf.setdefault(key, value) for key, value in add_setting_fwf.items()]

# <-1->    <setting_fwf> 'dtype' KEY INPUT STRING MODIFICATION
#          setting_fwf = {'dtype':'np.float64'}   =>   setting_fwf = {'dtype':'float64'}
        if type(setting_fwf['dtype']) == str:
            setting_fwf['dtype'] = setting_fwf['dtype'].replace('np.','')

# <-1->    <setting_fwf> 'dtype' KEY NUMPY VALUE ASSIGNMENT
#          setting_fwf = {'dtype': 'float64'}   =>   setting_fwf = {'dtype': np.float64}
            try:
                setting_fwf['dtype'] = getattr(np, setting_fwf['dtype'])
#          setting_fwf = {'dtype': 'float1000'}   =>   setting_fwf = {'dtype': None}
            except:
                setting_fwf['dtype'] = None

# <-1->    <setting_fwf> 'dtype' KEY INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'dtype': None}   =>   setting_fwf = {'dtype': 'float64'}
        if setting_fwf['dtype'] not in [np.float64, np.float32, np.float16, np.int64, np.int32, np.int16,\
                                    np.uint64, np.uint32, np.uint16, int, float]:
            setting_fwf['dtype'] = default_dtype
            logging.warning(f"Incorrectly defined <setting_fwf> dtype.")
            logging.warning(f"The <setting_fwf> dtype has been set to the default value.")
            warnings = True

# <-1->    <setting_fwf> 'dtype' KEY ASSIGNED VALUE DISPLAY
        logging.info(f"The <setting_fwf> dtype is set to: '{setting_fwf['dtype']}'")

# <-1->    <accepted_engine> VARIABLE CREATION
        accepted_engine = ['python-fwf', 'python', 'c', 'pyarrow']

# <-1->    <setting_fwf> 'engine' DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {}   =>   setting_fwf = {'engine': 'python'}
        if setting_fwf.get('engine') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_fwf['engine'] = 'python'
            logging.info(f"The <setting_fwf> engine has been set to the following default value: '{setting_fwf['engine']}'")

# <-1->    <setting_fwf> 'engine' CORRECT INPUT MESSAGE
        elif setting_fwf.get('engine') in accepted_engine:
            logging.info(f"The <setting_fwf> engine is set to: '{setting_fwf['engine']}'")

# <-1->    <setting_fwf> 'engine' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'engine': 'ABC'}   =>   setting_fwf = {'engine': 'python'}
        elif setting_fwf.get('engine') not in accepted_engine:
            setting_fwf['engine'] = 'python'
            logging.info(f"The accepted <setting_fwf> engine values are: {accepted_engine}")
            logging.warning(f"Incorrectly defined <setting_fwf> engine.")
            logging.warning(f"The <setting_fwf> engine has been set to the following default value: '{setting_fwf['engine']}'")
            warnings = True
    
# <-1->    <setting_fwf> 'skiprows' STRING TO INTEGER CONVERSION
#          setting_fwf = {'skiprows': '0'}   =>   setting_fwf = {'skiprows': 0}
        try:
            if type(setting_fwf['skiprows']) == str:
                setting_fwf['skiprows'] = int(setting_fwf['skiprows'])
#          setting_fwf = {'skiprows': 'one'}   <=>   setting_fwf = {'skiprows': 'one'}
        except:
            pass

# <-1->    <setting_fwf> 'skiprows' DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {}   =>   setting_fwf = {'skiprows': 0}
        if setting_fwf.get('skiprows') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_fwf['skiprows'] = 0
            logging.info(f"The <setting_fwf> skiprows has been set to the following default value: {setting_fwf['skiprows']}")
            logging.info(f"No initial rows have been skipped in the data loading process.")

# <-1->    <setting_fwf> 'skiprows' CORRECT INPUT MESSAGE
        elif type(setting_fwf.get('skiprows')) == int:
            logging.info(f"The <setting_fwf> skiprows is set to: {setting_fwf['skiprows']}")
            logging.info(f"The number of skipped initial rows in the data loading process was: {setting_fwf['skiprows']}")

# <-1->    <setting_fwf> 'skiprows' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'skiprows': 'one'}   =>   setting_fwf = {'skiprows': 0}
        elif type(setting_fwf.get('skiprows')) != int:
            setting_fwf['skiprows'] = 0
            logging.warning(f"The <setting_fwf> skiprows value must be an integer.")
            logging.warning(f"The <setting_fwf> skiprows has been set to the following default value: {setting_fwf['skiprows']}")
            logging.warning(f"No initial rows have been skipped in the data loading process.")
            warnings = True
    
# <-1->    <setting_fwf> 'skipfooter' STRING TO INTEGER CONVERSION
#          setting_fwf = {'skipfooter': '0'}   =>   setting_fwf = {'skipfooter': 0}
        try:
            if type(setting_fwf['skipfooter']) == str:
                setting_fwf['skipfooter'] = int(setting_fwf['skipfooter'])
#          setting_fwf = {'skipfooter': 'one'}   <=>   setting_fwf = {'skipfooter': 'one'}
        except:
            pass

# <-1->    <setting_fwf> 'skipfooter' DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {}   =>   setting_fwf = {'skipfooter': 0}
        if setting_fwf.get('skipfooter') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_fwf['skipfooter'] = 0
            logging.info(f"The <setting_fwf> skipfooter has been set to the following default value: {setting_fwf['skipfooter']}")
            logging.info(f"No end rows have been skipped in the data loading process.")

# <-1->    <setting_fwf> 'skipfooter' CORRECT INPUT MESSAGE
        elif type(setting_fwf.get('skipfooter')) == int:
            logging.info(f"The <setting_fwf> skipfooter is set to: {setting_fwf['skipfooter']}")
            logging.info(f"The number of skipped end rows in the data loading process was: {setting_fwf['skipfooter']}")

# <-1->    <setting_fwf> 'skipfooter' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'skipfooter': 'one'}   =>   setting_fwf = {'skipfooter': 0}
        elif type(setting_fwf.get('skipfooter')) != int:
            setting_fwf['skipfooter'] = 0
            logging.warning(f"The <setting_fwf> skipfooter value must be an integer.")
            logging.warning(f"The <setting_fwf> skipfooter has been set to the following default value: {setting_fwf['skipfooter']}")
            logging.warning(f"No end rows have been skipped in the data loading process.")
            warnings = True
    
# <-1->    <setting_fwf> 'displayed_decimals' STRING TO INTEGER CONVERSION
#          setting_fwf = {'displayed_decimals': '5'}   =>   setting_fwf = {'displayed_decimals': 5}
        try:
            if type(setting_fwf['displayed_decimals']) == str:
                setting_fwf['displayed_decimals'] = int(setting_fwf['displayed_decimals'])
#          setting_fwf = {'displayed_decimals': 'one'}   <=>   setting_fwf = {'displayed_decimals': 'one'}        
        except:
            pass

# <-1->    <setting_fwf> 'displayed_decimals' DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {}   =>   setting_fwf = {'displayed_decimals': 4}
        default_decimals = 4
        if setting_fwf.get('displayed_decimals') in [[''], [' '], [], (), {}, None, '', ' ']:
            setting_fwf['displayed_decimals'] = default_decimals
            logging.info(f"The <setting_fwf> displayed_decimals has been set to the following default value: {setting_fwf['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_fwf['displayed_decimals']}")

# <-1->    <setting_fwf> 'displayed_decimals' CORRECT INPUT MESSAGE
        elif type(setting_fwf.get('displayed_decimals')) == int:
            logging.info(f"The <setting_fwf> displayed_decimals is set to: {setting_fwf['displayed_decimals']}")
            logging.info(f"The number of displayed floating point decimals is: {setting_fwf['displayed_decimals']}")

# <-1->    <setting_fwf> 'displayed_decimals' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'displayed_decimals': 'one'}   =>   setting_fwf = {'displayed_decimals': 4}
        elif type(setting_fwf.get('displayed_decimals')) != int:
            setting_fwf['displayed_decimals'] = default_decimals
            logging.warning(f"The <setting_fwf> displayed_decimals value must be an integer.")
            logging.warning(f"The <setting_fwf> displayed_decimals has been set to the following default value: {setting_fwf['displayed_decimals']}")
            logging.warning(f"The number of displayed floating point decimals is: {setting_fwf['displayed_decimals']}")
            warnings = True
    
# <-1->    <accepted_encoding> VARIABLE CREATION
        accepted_encoding = ['utf-8', 'utf-16', 'utf-32', 'ascii']

# <-1->    <setting_fwf> 'encoding' DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {}   =>   setting_fwf = {'encoding': 'utf-8'}
        if setting_fwf.get('encoding') in [[''], [' '], [], (), {}, None, '', ' ']:
            add_setting_fwf = {'encoding':'utf-8'}
            setting_fwf['encoding'] = 'utf-8'
            logging.info(f"The <setting_fwf> encoding has been set to the following default value: '{setting_fwf['encoding']}'")

# <-1->    <setting_fwf> 'encoding' CORRECT INPUT MESSAGE
        elif setting_fwf.get('encoding') in accepted_encoding:
            logging.info(f"The <setting_fwf> encoding is set to: '{setting_fwf['encoding']}'")

# <-1->    <setting_fwf> 'encoding' INCORRECT INPUT DEFAULT VALUE ASSIGNMENT
#          setting_fwf = {'encoding': ''}   =>   setting_fwf = {'encoding': 'utf-8'}
        elif setting_fwf.get('encoding') not in accepted_encoding:
            setting_fwf['encoding'] = 'utf-8'
            logging.info(f"The accepted <setting_fwf> encoding values are: {accepted_encoding}")
            logging.warning(f"Incorrectly defined <setting_fwf> encoding.")
            logging.warning(f"The <setting_fwf> encoding has been set to the following default value: '{setting_fwf['encoding']}'")
            warnings = True

# <-1->    <setting_fwf> SETTING DISPLAY MESSAGE
        logging.info(f'The fwf read setting is set as follows: {list(setting_fwf.items())}')    
    
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    EXCEL DATA READ FUNCTION    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    

# <-0->    EXCEL DATA READ FUNCTION SECTION
    if file_format in [['.xlsx']] and data_read_option == ['pd.read_excel']:
        data = []
#          <data[i]> - STORES EXCEL VALUES OF <file[i]> PATH
        for index, file in enumerate(file_to_read):
            data.append(pd.read_excel(str(file[0]), sheet_name = setting_excel['sheet_name'], header = None, names = None, index_col = None, \
                usecols = setting_excel['usecols'], dtype = None, engine = setting_excel['engine'], converters = None, \
                skiprows = setting_excel['skiprows'], nrows = None, na_values = None, keep_default_na = True, \
                na_filter = True, thousands = None, skipfooter = setting_excel['skipfooter'], decimal = '.', comment = None, \
                storage_options = None, verbose = False, true_values = None, false_values = None))
        
#          <data> VARIABLE STRING TO INTEGER CONVERSION
#            data[index] = data[index].apply(pd.to_numeric, errors='coerce').astype(setting_excel['dtype']).round(setting_excel['displayed_decimals'])
#        logging.info(f"The data have been loaded. All non-numerical data have been converted to NaN values.")

#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    TABLE DATA READ FUNCTION    °°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°
#°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°°    

# <-0->    TABLE DATA READ FUNCTION SECTION
    elif file_format in [['.xlsx'], ['.csv'], ['.txt']] and data_read_option == ['pd.read_table']:
        data = []
#          <data[i]> - STORES FILE VALUES OF <file[i]> PATH
        for index,file in enumerate(file_to_read):
            data.append(pd.read_table(str(file[0]), header = None, names = None, index_col = None, usecols = setting_table['usecols'], dtype = None, \
                engine = setting_table['engine'], converters = None, skiprows = setting_table['skiprows'], skipfooter = setting_table['skipfooter'], \
                nrows = None, na_values = None, keep_default_na = True, na_filter = True, thousands = None, decimal = '.', comment = None, \
                storage_options = None, verbose = False, true_values = None, false_values = None, skip_blank_lines = True, parse_dates = False, \
                cache_dates = False, infer_datetime_format = False, keep_date_col = False, dayfirst = False, escapechar = None, \
                encoding = setting_table['encoding'], on_bad_lines = 'warn', skipinitialspace = True, delimiter = setting_table['delimiter'], \
                float_precision = None, iterator = False, chunksize = None, compression = 'infer', dialect = None, low_memory = True, \
                memory_map = False, encoding_errors = 'strict'))


```
